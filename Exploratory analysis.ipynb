```python
country = "Gambia"
country_data = df[df["Country"] == country]

values = [country_data["Criminality avg,"].values[0],
          country_data["Resilience avg,"].values[0]]

plt.bar(["Criminality", "Resilience"], values, color=["red", "green"])
plt.title(f"{country}: Criminality vs Resilience")
plt.ylabel("Average Value")
plt.show()
```

```python
# STEP 6: Threshold Categorization
# ---------------------------------------
def cat_criminality(x):
    if x <= 2: return "Minimal"
    if x <= 5: return "Moderate"
    if x <= 7: return "Significant"
    return "Severe"

def cat_resilience(x):
    if x <= 3: return "Weak"
    if x <= 5: return "Moderate"
    if x <= 7: return "Sufficient"
    return "Highly_effective"

df['Criminality Category'] = df['Criminality avg,'].apply(cat_criminality)
df['Resilience Category'] = df['Resilience avg,'].apply(cat_resilience)

print(df[['Country','Criminality avg,','Criminality Category']])

# Countplot
sns.countplot(x="Criminality Category", data=df)
plt.title("Criminality Categories")
plt.show()
```

```python
# 🔍 Filter for only Gambia
gambia_info = df[df['Country'] == "Gambia"][[
    'Country',
    'Criminality avg,',
    'Criminality Category',
    'Resilience avg,',
    'Resilience Category',
    'Human trafficking'
]]

print("Threshold categories for The Gambia:\n", gambia_info)
```
